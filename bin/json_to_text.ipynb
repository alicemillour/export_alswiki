```python
# create one text file from the result obtained with https://github.com/attardi/wikiextractor
# when executed on a Wikipedia dump

import json

elsassisch_not_ort='../elsassisch_not_ort/AA/wiki_00'
elsassisch_file_00='../elsassisch/AA/wiki_00'
elsassisch_file_01='../elsassisch/AA/wiki_01'

def json_to_text_file(json_file, text_file):
    with open(json_file) as json_file:
        f = open(text_file, "a")
        for article in json_file:
            data = json.loads(article)
            print(data['text'])
            f.write(data['text'])
        f.close()
        
json_to_text_file(elsassisch_not_ort,"../corpora/elsassisch_not_ort.txt")
json_to_text_file(elsassisch_file_00,"../corpora/elsassisch.txt")
json_to_text_file(elsassisch_file_01,"../corpora/elsassisch.txt")
```
